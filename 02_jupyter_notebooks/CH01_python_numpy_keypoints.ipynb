2. String Operation

```python
s = "hello"
print(s.capitalize())  # Capitalize a string
print(s.upper())       # Convert a string to uppercase; prints "HELLO"
print(s.rjust(7))      # Right-justify a string, padding with spaces
print(s.center(7))     # Center a string, padding with spaces
print(s.replace('l', '(ell)'))  # Replace all instances of one substring with another
print('  world '.strip())
```

3. List

```python
# list creation
list_1 = ['one', 'two', 'three']
list_1.append(4)
list_1.insert(0, 'ZERO')

list_2 = [1, 2, 3]
list_1.extend(list_2)
print(list_1)
```

```python
# list comprehension
long_list = [i for i in range(9)]
long_long_list = [(i, j) for i in range(3) for j in range(5)]
print(long_long_list)
```

```python
# list comprehension with conditional
conditional_list1 = [i**2 for i in range(10) if i%2==0]
print(conditional_list1)
conditional_list2 = [i**2 if i%2==0 else i**3 for i in range(10) ]
print(conditional_list2)
```

```python
# list sort
random_list_2 = [(3, 'z'), (12, 'r'), (6, 'e'),(8, 'c'), (2, 'g')]
sorted(random_list_2, key=lambda x: x[1])
```

4. Dict and Set

```python
# set
animals = {'cat', 'dog'}
print('cat' in animals)   # Check if an element is in a set; prints "True"
print('fish' in animals)  # prints "False"

animals.add('cat')       # Adding an element that is already in the set does nothing
print(len(animals))
animals.remove('cat')    # Remove an element from a set
print(len(animals))

animals.add('fish')

for idx, animal in enumerate(animals):
    print('#{}: {}'.format(idx + 1, animal))
```

```python
# dict
d = {'cat': 'cute', 'dog': 'furry'}  # Create a new dictionary with some data
print(d['cat'])       # Get an entry from a dictionary; prints "cute"
print('cat' in d)     # Check if a dictionary has a given key; prints "True"

d['fish'] = 'wet'    # Set an entry in a dictionary
print(d['fish'])      # Prints "wet"

print(d.get('monkey', 'N/A'))  # Get an element with a default; prints "N/A"
print(d.get('fish', 'N/A'))    # Get an element with a default; prints "wet"

del d['fish']        # Remove an element from a dictionary
print(d.get('fish', 'N/A')) # "fish" is no longer a key; prints "N/A"
```

```python
my_dict = {(5 - i): i ** 2 for i in range(10)}
second_dict = {'a': 10, 'b': 11}
my_dict.update(second_dict)

for k, it in my_dict.items():
  print(k, it)
```

```python
# Tuple
d = {(x, x + 1): x for x in range(10)}  # Create a dictionary with tuple keys
t = (5, 6)       # Create a tuple
print(type(t))
print(d)
print(d[t])
print(d[(1, 2)])
```

Numpy

matrix multiplication

```python
import numpy as np

# initialize the random weight matrix W and bias array B
W = np.random.random((4,3)) # Create an array filled with random values
# W = np.random.random(4,3)
print(W)
B = np.array([0.1, 0.2, 0.3, 0.4]).T
print(B)
```

```python
# Y = W*X + B
X = np.array([1, 2, 3])
W_X = np.matmul(W, X.T)
print(W_X)
Y = W_X + B
print(Y)
```

dot product /elementary wise multiplication

```python
# elementary wise
a = np.array([i for i in range(10)]).reshape(2, 5)
print(a)
print(a * a)
np.multiply(a, a)
np.multiply(a, 10)
```

```python
# dot product between vectors or between vector and matrix
v = np.array([9,10])
w = np.array([11, 12])

# Inner product of vectors; both produce 219
print(v.dot(w))
print(np.dot(v, w))
print(v @ w)
```

```python
x = np.array([[1,2],[3,4]])
print(x.dot(v))
print(np.dot(x, v))
print(x @ v)
```

numpy sum

```python
x = np.array([[1,2],[3,4]])

print(np.sum(x))  # Compute sum of all elements; prints "10"
print(np.sum(x, axis=0))  # Compute sum of each column; prints "[4 6]"
print(np.sum(x, axis=1))  # Compute sum of each row; prints "[3 7]"
```

broadcasting

```python
samples = np.random.random((15, 5))
# Without broadcasting
expanded1 = np.expand_dims(samples, axis=1)
print(expanded1.shape)
tile1 = np.tile(expanded1, (1, samples.shape[0], 1))
print(tile1.shape)
expanded2 = np.expand_dims(samples, axis=0)
print(expanded2.shape)
tile2 = np.tile(expanded2, (samples.shape[0], 1 ,1))
print(tile2.shape)
diff = tile2 - tile1
print(diff.shape)
distances = np.linalg.norm(diff, axis=-1)
print(distances.shape)
```

Matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# Setup grid with height 2 and col 1.
# Plot the 1st subplot
plt.subplot(2, 1, 1)
plt.grid()
plt.plot(x, y_sin)
plt.title('Sine Wave')
# Now plot on the 2nd subplot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine Wave')
plt.grid()
plt.tight_layout()
```
